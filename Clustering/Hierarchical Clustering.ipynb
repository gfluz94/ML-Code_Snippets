## Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

## Dataset

```python
df = pd.read_csv("Mall_Customers.csv", index_col="CustomerID")
```

```python
quantitative = [column for column in df.columns if df[column].dtype!="object"]
qualitative = [column for column in df.columns if column not in quantitative]
```

## Clustering reduced dataset

```python
X = df.loc[:, quantitative[1:]]
```

```python
fig, ax = plt.subplots(1,1,figsize=(15,6))
dendogram = sch.dendrogram(sch.linkage(X, method="ward"), ax=ax)
```

```python
optimal_clusters = 5

aggCluster = AgglomerativeClustering(n_clusters=optimal_clusters, affinity='euclidean', linkage='ward')
clusters = aggCluster.fit_predict(X)
```

```python
import seaborn as sns

labels = ["Careful", "Standard", "Target", "Careless", "Sensible"]
new_clusters = map(lambda x: labels[x], clusters)

fig, ax = plt.subplots(1,1, figsize=(10,6))
sns.scatterplot(x=X.iloc[:,-2], y=X.iloc[:,-1], hue=new_clusters, 
                palette=["red", "blue", "green", "cyan", "magenta"], ax=ax)
plt.title(f"K-Means Clustering with K={optimal_clusters}")
plt.show()
```

## Clustering complete dataset

```python
X = df.loc[:, :]
```

```python
class ColumnSelector(BaseEstimator, TransformerMixin):
    
    def __init__(self, columns):
        self.columns = columns
        
    def fit(self, data, labels=None):
        return self
    
    def transform(self, data):
        assert isinstance(data, pd.DataFrame), "ERROR: Insert pandas DataFrame"
        
        return data[self.columns]
    
class ArrayConverter(BaseEstimator, TransformerMixin):
    
    def __init__(self):
        pass
        
    def fit(self, data, labels=None):
        return self
    
    def transform(self, data):        
        return data.toarray()
```

```python
def build_pipeline(model):
    categorical = Pipeline(steps=[
        ("cat_selection", ColumnSelector(columns=qualitative)),
        ("cat_onehot", OneHotEncoder()),
        ("to_array", ArrayConverter())
    ])
    numerical = Pipeline(steps=[
        ("num_selection", ColumnSelector(columns=quantitative)),
        ("num_onehot", StandardScaler())
    ])
    preprocessor = FeatureUnion(transformer_list=[("categorical", categorical), ("numerical", numerical)])
    pipe = Pipeline(steps=[
        ("preprocessor", preprocessor),
        ("model", model)
    ])
    return pipe
```

```python
fig, ax = plt.subplots(1,1,figsize=(15,6))
aggCluster = AgglomerativeClustering()
model = build_pipeline(aggCluster)
dendogram = sch.dendrogram(sch.linkage(model[0].fit_transform(X), method="ward"), ax=ax)
```

```python
optimal_clusters = 3

aggCluster = AgglomerativeClustering(n_clusters=optimal_clusters, affinity='euclidean', linkage='ward')
model = build_pipeline(aggCluster)
clusters = model.fit_predict(X)
```

```python
import seaborn as sns

labels = ["Target", "Standard", "Careful"]
new_clusters = map(lambda x: labels[x], clusters)

fig, ax = plt.subplots(1,1, figsize=(10,6))
sns.scatterplot(x=X.iloc[:,-2], y=X.iloc[:,-1], hue=new_clusters, 
                palette=["red", "blue", "green"], ax=ax)
plt.title(f"K-Means Clustering with K={optimal_clusters}")
plt.show()
```
