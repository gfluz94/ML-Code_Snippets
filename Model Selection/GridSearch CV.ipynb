## Libraries

## Splitting dataset

```python
X = df.iloc[:, 1:-1]
y = df.iloc[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

## Complete Model

```python
class ColumnSelector(BaseEstimator, TransformerMixin):
    
    def __init__(self, columns):
        self.columns = columns
        
    def fit(self, data, labels=None):
        return self
    
    def transform(self, data):
        assert isinstance(data, pd.DataFrame), "Insert a pandas dataframe."
        
        return data[self.columns]
```

```python
num_preprocessor = Pipeline(steps=[
    ("num_selector", ColumnSelector(columns=quantitative)),
    ("standardization", StandardScaler())])
cat_preprocessor = Pipeline(steps=[
    ("cat_selector", ColumnSelector(columns=qualitative)),
    ("onehot", OneHotEncoder())])
preprocessor = FeatureUnion(transformer_list=[("categorical", cat_preprocessor), ("numerical", num_preprocessor)])

model = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("svm", SVC(probability=True, random_state=0))])
```

+ Searching for the best hyperparameters

```python
params = [{
           "svm__C": [0.01, 0.03, 0.1, 0.3, 1, 3, 10],
           "svm__kernel": ["linear"]},
          {
           "svm__C": [0.01, 0.03, 0.1, 0.3, 1, 3, 10],
           "svm__kernel": ["rbf"],
           "svm__gamma": ["scale", 0.01, 0.03, 0.1, 0.3, 1, 3, 10]}]

grid_search = GridSearchCV(estimator=model, param_grid=params, scoring="accuracy", cv=5,
                           refit=True, n_jobs=-1, verbose=1)
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

+ Applying best model (**refit=True** implies training model on whole dataset with the best found hyperparameters)

```python
y_pred = grid_search.predict(X_test)
```

+ Metrics

```python
print(classification_report(y_test, y_pred))
```

```python
y_predicted = pd.DataFrame(data=y_pred, index=y_test.index, columns=["Predicted"])
complete = pd.merge(left=y_test, right=y_predicted, right_index=True, left_index=True)

complete["Purchased"] = complete["Purchased"].apply(lambda x: "Yes" if x==1 else "No")
complete["Predicted"] = complete["Predicted"].apply(lambda x: "Yes" if x==1 else "No")

conf_matrix = complete.groupby(["Purchased", "Predicted"]).size().unstack("Predicted")
conf_matrix
```

```python
probabilities = grid_search.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, probabilities)

fig, ax = plt.subplots(1,1, figsize=(10,6))
plt.plot(fpr, tpr, color="red", label=f"RBF-Kernel SVC (AUC: {roc_auc_score(y_test, probabilities):.3f})")
plt.plot([0,1], [0,1], color="navy")
plt.xlabel("FPR", fontsize=12)
plt.ylabel("TPR", fontsize=12)
plt.xlim((0,1))
plt.ylim((0,1.001))
plt.legend(loc=4, fontsize=10)
plt.title("ROC-AUC Analysis", fontsize=13)
plt.show()
```

+ Visualizing Decision Boundary

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2, random_state=0)

pca.fit(grid_search.best_estimator_[0].transform(X_train).toarray())
X_pca = pca.transform(grid_search.best_estimator_[0].transform(X_test).toarray())
```

```python
from matplotlib.colors import ListedColormap

fig, ax = plt.subplots(1,1,figsize=(12,8))
sns.scatterplot(X_pca[:,0], X_pca[:,1], 
                    hue=y_test, palette=["navy", "red"], ax=ax)
xx, yy = np.mgrid[X_pca[:,0].min()-1:X_pca[:,0].max()+1:0.01,
                  X_pca[:,1].min()-1:X_pca[:,1].max()+1:0.01]
grid = np.c_[xx.ravel(), yy.ravel()]
classes = grid_search.best_estimator_[-1].predict(pca.inverse_transform(grid)).reshape(xx.shape)
ax.contourf(xx, yy, classes, cmap=ListedColormap(("navy", "red")), alpha=0.2)
ax.set_title("Test Set - Decision Boundary")
ax.set_xlabel("PCA1")
ax.set_ylabel("PCA2")
plt.show()
```
